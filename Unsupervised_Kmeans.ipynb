```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

X = np.random.normal(1,1,[100,2]) #são esses os dados gerados das distribuições, para simular um overlap, basta reduzir a média.
Y = np.random.normal(3,1,[100,2])

plt.scatter(X[:,0],X[:,1])
plt.scatter(Y[:,0],Y[:,1])
plt.show()
target_1 = ['azul' for x in range(100)]
target_2 = ['laranja' for x in range(100)]
my_dc = {'atributo 1':X[:,0],'atributo 2':X[:,1],'target':target_1}
my_dc2= {'atributo 1':Y[:,0],'atributo 2':Y[:,1],'target':target_2}

DF = pd.DataFrame(my_dc)
DF2 = pd.DataFrame(my_dc2)

DF=pd.concat([DF,DF2],axis=0)



centroide1=np.random.rand(2)
centroide2 = np.random.rand(2)

def euclideanDist(X,Y):
  return np.sqrt((X[0]-Y[0])**2+(X[1]-Y[1])**2)
dist1=[]
dist2=[]
iterations=100

cluster1=[]
cluster2=[]

for i in range(iterations):
  print('iteracao %d'%(i))
  print(centroide1)
  print(centroide2)
  dist1=[]
  dist2=[]
  cluster1=[]
  cluster2=[]
  for j in range(200):
    dist1.append(euclideanDist(centroide1,DF.iloc[j,0:2].tolist()))
    dist2.append(euclideanDist(centroide2,DF.iloc[j,0:2].tolist()))

    if dist1[j]>dist2[j]:
      cluster2.append(DF.iloc[j,0:2].tolist())
    else:
      cluster1.append(DF.iloc[j,0:2].tolist())
  print(len(cluster1))
  print(len(cluster2))

  centroide1=[np.mean(np.array(cluster1)[:,0]),np.mean(np.array(cluster1)[:,1])]
  centroide2=[np.mean(np.array(cluster2)[:,0]),np.mean(np.array(cluster2)[:,1])]











```
