```python

import numpy as np
from matplotlib import pyplot as plt

x = np.arange(0,2*np.pi,0.1)   # start,stop,step
y = np.sin(x)  # funo seno



xn = np.arange(0,2*np.pi,2*np.pi/10) # Gerando 10 pontos entre 0 e 2*pi*x
yn = np.sin(xn) + np.random.normal(0,0.5,10) # adio de rudo

plt.plot(x,y) # plot de dados originais    - linha
plt.plot(xn,yn,'o') # plot de dados ruidosos - pontos
plt.legend(['Real','Dados Ruidosos'])
plt.title('Funo seno e dados ruidosos')
plt.show()
maxOrder = 10  # mxima ordem do polinmio

poly = []
polysin = []
for M in range(maxOrder):

    poly=[]
    polysin =[]
    for i in range(len(xn)):
        poly.append( xn[i]**np.arange(0,M+1,1))
    for i in range(len(x)):
        polysin.append(x[i]**np.arange(0,M+1,1))
    poly_matrix = np.matrix(poly)
    polysin_matrix = np.matrix(polysin)
    coefs = np.linalg.pinv(poly_matrix).dot(yn.T) # mnimos quadrados (least squares)
    outputs = polysin_matrix.dot(coefs.T)
    plt.plot(x,y)
    plt.plot(xn,yn,'o')
    plt.plot(x,outputs)
    plt.legend(['Real','Dados Ruidosos','Estimativas'])
    plt.title('Estimativas utilizando polinmio de ordem M = %d'%(M))
    plt.show()
```
