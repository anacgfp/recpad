```python
import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error as mse
from matplotlib import pyplot as plt

#3.5*x + 2*y
matrix_input = np.array([[1,2],[2,2],[3,4],[5,6],[5,10]])
output = np.array([4.814213562373095,
 11.514213562373095,
 23.93333333333333,
 62.64948974278318,
 63.36227766016838])

iterations = 10000
n = 0.001

theta = np.random.rand(3)
theta_zero = np.random.rand()
mses = []
preds=[]
for i in range(iterations):
  preds=[]
  for j in range(len(matrix_input)):
    preds.append(theta[0]*matrix_input[j][0]**2 + theta[1]*matrix_input[j][1]**0.5 +theta[2]*matrix_input[j][0] +theta_zero)

    erro= (output[j] - preds[j])

    theta = theta + n*erro*matrix_input[j]
    theta_zero = theta_zero + n*erro
  mses.append(mse(output,preds))

plt.plot(mses)
plt.xlabel='Iterations'
plt.ylabel='MSE'
plt.show()







```
